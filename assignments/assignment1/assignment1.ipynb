### task 1

```python
# Original list of transactions
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

# Step a: filter out invalid transactions
valid_transactions = []
for t in transactions:
    if t["price"] > 0 and t["quantity"] >= 0:
        valid_transactions.append(t)

# Step b: calculate total for each transaction
for t in valid_transactions:
    t["total"] = t["price"] * t["quantity"]

# Step c: group by region and sum revenue
revenue_by_region = {}

for t in valid_transactions:
    region = t["region"]
    total = t["total"]

    if region in revenue_by_region:
        revenue_by_region[region] += total
    else:
        revenue_by_region[region] = total

# Step d: find region with highest revenue
best_region = None
highest_value = 0

for region in revenue_by_region:
    if revenue_by_region[region] > highest_value:
        highest_value = revenue_by_region[region]
        best_region = region

print(f"Top region: {best_region} with revenue {highest_value}")
```

### task 2

```python
# List of daily sales
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(sales):
    # Calculate total sales
    total = 0
    for s in sales:
        total += s

    # Calculate average
    average = round(total / len(sales), 2)

    # Find best and worst day
    highest = sales[0]
    lowest = sales[0]
    best_day = 1
    worst_day = 1

    for i in range(len(sales)):
        if sales[i] > highest:
            highest = sales[i]
            best_day = i + 1

        if sales[i] < lowest:
            lowest = sales[i]
            worst_day = i + 1

    # Count days above average
    count = 0
    for s in sales:
        if s > average:
            count += 1

    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": count
    }

# Call function
result = summarise_sales(sales)
print(result)
```

### task 3

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

# a) students who passed
passing = []
for s in students:
    if s["grade"] >= 50:
        passing.append(s["name"])

print(passing)

# b) print status for each student
for s in students:
    if s["grade"] >= 80:
        print(f"{s['name']}: Excellent")
    elif s["grade"] >= 50:
        print(f"{s['name']}: Pass")
    else:
        print(f"{s['name']}: Fail")

# c) count low attendance
count = 0
for s in students:
    if s["attendance"] < 0.80:
        count += 1

print("Low attendance:", count)

# d) find top student using loop
top_student = students[0]

for s in students:
    if s["grade"] > top_student["grade"]:
        top_student = s

print("Top student:", top_student["name"], top_student["grade"])

# e) using max()
top_student2 = max(students, key=lambda s: s["grade"])
print("Top student (max):", top_student2["name"], top_student2["grade"])
```

### task 4

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}

    # Add or update asset
    def add_asset(self, ticker, shares, price_per_share):
        self.holdings[ticker] = {
            "shares": shares,
            "price_per_share": price_per_share
        }

    # Calculate total portfolio value
    def total_value(self):
        total = 0
        for ticker in self.holdings:
            shares = self.holdings[ticker]["shares"]
            price = self.holdings[ticker]["price_per_share"]
            total += shares * price
        return total

    # Find most valuable asset
    def most_valuable(self):
        best_ticker = None
        best_value = 0

        for ticker in self.holdings:
            shares = self.holdings[ticker]["shares"]
            price = self.holdings[ticker]["price_per_share"]
            value = shares * price

            if value > best_value:
                best_value = value
                best_ticker = ticker

        return best_ticker

    # String representation
    def __str__(self):
        return f"Owner: {self.owner}, Assets: {len(self.holdings)}, Total value: {self.total_value()}"

# Create portfolio
p = Portfolio("Alice")

# Add assets
p.add_asset("AAPL", 10, 150)
p.add_asset("TSLA", 5, 700)
p.add_asset("GOOG", 2, 2800)

# Show results
print(p)
print("Total value:", p.total_value())
print("Most valuable asset:", p.most_valuable())
```

### task 5

```python
from datetime import datetime

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

# Convert strings to date objects
dates = []

for order in orders:
    date_string = order[0]
    date_object = datetime.strptime(date_string, "%Y-%m-%d").date()
    dates.append(date_object)

# Calculate difference
difference = max(dates) - min(dates)

print("Days between earliest and latest:", difference.days)
```

### task 8

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)

days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

# create dataframe
df = pd.DataFrame({
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
})

df = df.set_index("date")

# new columns
df["conversions"] = round(df["visitors"] * df["conversion_rate"])
df["daily_revenue"] = df["conversions"] * df["avg_order_value"]

print(df.head())
```
