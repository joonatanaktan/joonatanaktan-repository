The dataset used in this project is the Austin Weather Dataset. It contains daily weather observations recorded in Austin, Texas.

The dataset includes variables such as:
Date
Temperature (high, average, low)
Dew point (high, average, low)
Humidity (high, average, low)
Sea level pressure (high, average, low)
Visibility (high, average, low)
Wind speed (high, average, gust)
Precipitation
Weather events (e.g., rain, thunderstorms)

The dataset contains 1319 rows and 21 columns, representing daily observations over several years.

In this project, I will use the dataset to perform exploratory data analysis, identify patterns in weather over time, and examine relationships between variables such as temperature, humidity, and precipitation.

```python
# C
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

weather_df = pd.read_csv("austin_weather.csv")

weather_df["Date"] = pd.to_datetime(weather_df["Date"])

numeric_columns = [
    "TempHighF", "TempAvgF", "TempLowF",
    "DewPointHighF", "DewPointAvgF", "DewPointLowF",
    "HumidityHighPercent", "HumidityAvgPercent", "HumidityLowPercent",
    "SeaLevelPressureHighInches", "SeaLevelPressureAvgInches", "SeaLevelPressureLowInches",
    "VisibilityHighMiles", "VisibilityAvgMiles", "VisibilityLowMiles",
    "WindHighMPH", "WindAvgMPH", "WindGustMPH"
]

for col in numeric_columns:
    weather_df[col] = pd.to_numeric(weather_df[col], errors="coerce")

weather_df["PrecipitationSumInches"] = pd.to_numeric(
    weather_df["PrecipitationSumInches"], errors="coerce"
)

print("First 5 rows:")
display(weather_df.head())

print("Summary statistics:")
display(weather_df.describe())

print("Missing values:")
display(weather_df.isnull().sum())

plt.figure(figsize=(12, 5))
sns.lineplot(data=weather_df, x="Date", y="TempAvgF")
plt.title("Average Temperature Over Time in Austin")
plt.xlabel("Date")
plt.ylabel("Average Temperature (°F)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(weather_df["HumidityAvgPercent"].dropna(), bins=30, kde=True)
plt.title("Distribution of Average Humidity")
plt.xlabel("Average Humidity (%)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
import sqlite3
import pandas as pd

# connect to database
conn = sqlite3.connect("northwind.db")

# a) list all tables
tables = pd.read_sql_query(
    "SELECT name FROM sqlite_master WHERE type='table';",
    conn
)

print("Tables in the database:")
display(tables)

# b) products with UnitPrice > 30
query_b = """
SELECT ProductName, UnitPrice, UnitsInStock
FROM Products
WHERE UnitPrice > 30;
"""

df_products = pd.read_sql_query(query_b, conn)

print("\nProducts with UnitPrice > 30:")
display(df_products)

# c) orders shipped to Germany
query_c = """
SELECT c.CompanyName,
       o.OrderID,
       o.OrderDate,
       o.ShipCountry,
       (od.UnitPrice * od.Quantity) AS LineTotal
FROM Orders o
JOIN Customers c
    ON o.CustomerID = c.CustomerID
JOIN "Order Details" od
    ON o.OrderID = od.OrderID
WHERE o.ShipCountry = 'Germany';
"""

df_germany = pd.read_sql_query(query_c, conn)

print("\nOrders shipped to Germany:")
display(df_germany)

conn.close()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# a) Top 10 customers by number of orders
orders_per_customer = (
    df_germany[["CompanyName", "OrderID"]]
    .drop_duplicates()
    .groupby("CompanyName")
    .size()
    .sort_values(ascending=False)
    .head(10)
    .reset_index(name="NumberOfOrders")
)

plt.figure(figsize=(10, 6))
sns.barplot(data=orders_per_customer, x="NumberOfOrders", y="CompanyName")
plt.title("Top 10 Customers by Number of Orders Shipped to Germany")
plt.xlabel("Number of Orders")
plt.ylabel("Customer")
plt.tight_layout()
plt.show()

# b) KDE plot of daily total sales
df_germany["OrderDate"] = pd.to_datetime(df_germany["OrderDate"], format="mixed", errors="coerce")
df_germany["OrderDateOnly"] = df_germany["OrderDate"].dt.date

daily_sales = (
    df_germany.groupby("OrderDateOnly")["LineTotal"]
    .sum()
    .reset_index()
)

plt.figure(figsize=(8, 5))
sns.kdeplot(data=daily_sales, x="LineTotal", fill=True)
plt.title("Density Plot of Daily Total Sales (Germany Orders)")
plt.xlabel("Daily Total Sales")
plt.ylabel("Density")
plt.tight_layout()
plt.show()

# c) SQL query: total revenue per category
import sqlite3
conn = sqlite3.connect("northwind.db")

query_6c = """
SELECT c.CategoryName,
       SUM(od.UnitPrice * od.Quantity) AS TotalRevenue
FROM "Order Details" od
JOIN Products p ON od.ProductID = p.ProductID
JOIN Categories c ON p.CategoryID = c.CategoryID
GROUP BY c.CategoryName
ORDER BY TotalRevenue DESC;
"""

df_category_revenue = pd.read_sql_query(query_6c, conn)

print("Total revenue by category:")
display(df_category_revenue)

plt.figure(figsize=(10, 6))
sns.barplot(data=df_category_revenue, x="TotalRevenue", y="CategoryName")
plt.title("Total Revenue by Product Category")
plt.xlabel("Total Revenue")
plt.ylabel("Category")
plt.tight_layout()
plt.show()

conn.close()
```

```python
import sqlite3
import pandas as pd
import time

conn = sqlite3.connect("northwind.db")

# 7a) retrieve tables
products_df = pd.read_sql_query("SELECT * FROM Products;", conn)
categories_df = pd.read_sql_query("SELECT * FROM Categories;", conn)

print("Products (first 5 rows):")
display(products_df.head())

print("Categories (first 5 rows):")
display(categories_df.head())

# 7b) inner join
products_categories_joined = pd.merge(
    products_df,
    categories_df,
    on="CategoryID",
    how="inner"
)

print("Joined table (first 5 rows):")
display(products_categories_joined.head())

# 7c) SQL version timing
start_sql = time.time()

query_sql = """
SELECT c.CategoryName,
       SUM(od.UnitPrice * od.Quantity) AS TotalRevenue
FROM "Order Details" od
JOIN Products p ON od.ProductID = p.ProductID
JOIN Categories c ON p.CategoryID = c.CategoryID
GROUP BY c.CategoryName
ORDER BY TotalRevenue DESC;
"""

sql_result_df = pd.read_sql_query(query_sql, conn)

end_sql = time.time()
sql_time = end_sql - start_sql

# pandas version timing
start_python = time.time()

order_details_df = pd.read_sql_query('SELECT * FROM "Order Details";', conn)
products_df_full = pd.read_sql_query("SELECT * FROM Products;", conn)
categories_df_full = pd.read_sql_query("SELECT * FROM Categories;", conn)

merged_df = pd.merge(order_details_df, products_df_full, on="ProductID", how="inner")
merged_df = pd.merge(merged_df, categories_df_full, on="CategoryID", how="inner")

# clean calculation (no warning)
python_result_df = (
    merged_df.assign(Total=merged_df["UnitPrice_x"] * merged_df["Quantity"])
    .groupby("CategoryName")["Total"]
    .sum()
    .reset_index(name="TotalRevenue")
    .sort_values("TotalRevenue", ascending=False)
    .reset_index(drop=True)
)

end_python = time.time()
python_time = end_python - start_python

print("SQL result:")
display(sql_result_df)

print("Python/pandas result:")
display(python_result_df)

print(f"SQL execution time: {sql_time:.6f} seconds")
print(f"Python execution time: {python_time:.6f} seconds")

conn.close()
```

```python
import pandas as pd

# original wide data
wide_data = pd.DataFrame({
    "store": ["Helsinki", "Tampere", "Turku"],
    "Q1_sales": [12000, 8500, 9200],
    "Q2_sales": [13500, 9100, 9800],
    "Q3_sales": [11800, 8800, 10100],
    "Q4_sales": [15000, 10200, 11500]
})

# original long data
long_data = pd.DataFrame({
    "month": ["Jan", "Jan", "Feb", "Feb", "Mar", "Mar"],
    "product": ["Laptop", "Phone", "Laptop", "Phone", "Laptop", "Phone"],
    "units_sold": [45, 120, 50, 110, 55, 130]
})

# a) wide to long
print("Original wide_data:")
display(wide_data)

wide_to_long = pd.melt(
    wide_data,
    id_vars="store",
    var_name="quarter",
    value_name="sales"
)

print("wide_data converted to long format:")
display(wide_to_long)

# b) long to wide
print("Original long_data:")
display(long_data)

long_to_wide = long_data.pivot(
    index="month",
    columns="product",
    values="units_sold"
).reset_index()

print("long_data converted to wide format:")
display(long_to_wide)

# c) convert both back to original format

# long back to wide
restored_wide = wide_to_long.pivot(
    index="store",
    columns="quarter",
    values="sales"
).reset_index()

restored_wide.columns.name = None
restored_wide = restored_wide[["store", "Q1_sales", "Q2_sales", "Q3_sales", "Q4_sales"]]

print("Restored wide_data:")
display(restored_wide)

# wide back to long
restored_long = long_to_wide.melt(
    id_vars="month",
    var_name="product",
    value_name="units_sold"
).sort_values(["month", "product"]).reset_index(drop=True)

original_long_sorted = long_data.sort_values(["month", "product"]).reset_index(drop=True)

print("Original long_data sorted:")
display(original_long_sorted)

print("Restored long_data:")
display(restored_long)
```
